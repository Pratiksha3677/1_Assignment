#Q11. Statistical Analysis with NumPy

1. Calculate the total score for each student (sum of AssignmentScore, MidtermScore, FinalScore, ProjectScore)
2. Use NumPy to find students whose total score is above the 80th percentile
3. Calculate the standard deviation of AttendancePercent using NumPy

```python
#Calculate the total score for each student (sum of AssignmentScore, MidtermScore, FinalScore, ProjectScore)
import numpy as np
total_score = (
    data['AssignmentScore'].to_numpy()
    + data['MidtermScore'].to_numpy()
    + data['FinalScore'].to_numpy()
    + data['ProjectScore'].to_numpy()
)
data['TotalScore'] = total_score
print(total_score)
next_line()
#Use NumPy to find students whose total score is above the 80th percentile
percentile_80 = np.percentile(total_score, 80)
high_total_students = data[total_score > percentile_80]
print("80th Percentile Total Score:", percentile_80)
next_line()
#Calculate the standard deviation of AttendancePercent using NumPy
attendance_array = data['AttendancePercent'].to_numpy()
std_attendance = np.std(attendance_array)
print("Standard Deviation of Attendance:", std_attendance)
```

#Q12. Feature Engineering

1. Create a new column 'TotalScore' by adding all score columns
2. Create a 'Grade' column based on CGPA: A (3.5-4.0), B (3.0-3.5), C (2.5-3.0), D (below 2.5)
3. Create a 'PassStatus' column: 'Pass' if FinalScore >= 50, else 'Fail'
4. Calculate the percentage of students who passed
Export the modified DataFrame to a new CSV file called 'student_data_processed.csv'

```python
import numpy as np
data['TotalScore'] = (
    data['AssignmentScore'] +
    data['MidtermScore'] +
    data['FinalScore'] +
    data['ProjectScore']
)
data['Grade'] = np.where(data['CGPA'] >= 3.5, 'A',
                 np.where(data['CGPA'] >= 3.0, 'B',
                 np.where(data['CGPA'] >= 2.5, 'C', 'D')))
data['PassStatus'] = np.where(data['FinalScore'] >= 50, 'Pass', 'Fail')
pass_percentage = (data['PassStatus'] == 'Pass').mean() * 100
print("Percentage of Students who Passed:", pass_percentage)
data.to_csv('student_data_processed.csv', index=False)
```
