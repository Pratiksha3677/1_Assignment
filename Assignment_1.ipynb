```python
#Python Program to find the largest element in the list

numbers = [10, 25, 47, 3, 98, 56]

largest = max(numbers)

print("Largest element:", largest)
```

```python
#without using max function
numbers = [10, 25, 47, 3, 98, 56]

largest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num

print("Largest element:", largest)
```

```python
#using sort function
numbers = [10, 25, 47, 3, 98, 56]

numbers.sort()

print("Largest element:", numbers[-1])
```

```python
#Python program to find the sum of elements in a list
lst = [10, 20, 30, 40, 50]

total = sum(lst)

print("Sum of elements:", total)
```

```python
#Python program to Multiply all numbers in the list
lst = [2, 3, 4, 5]

product = 1
i = 0

while i < len(lst):
    product *= lst[i]
    i += 1

print("Product of elements:", product)
```

```python
#Python program to find the smallest and largest number in a list (Without min-max function)

lst = [10, 25, 3, 99, 45, 1, 67]
smallest = lst[0]
largest = lst[0]

for num in lst:
    if num < smallest:
        smallest = num
    if num > largest:
        largest = num

print("Smallest number:", smallest)
print("Largest number:", largest)
```

```python
#Take inputs from the user to make a list. Again take one input from the user and search it in the list and delete that element, if found. Iterate over a list using for loop.

# Step 1: Take inputs from the user to create a list
n = int(input("How many elements you want to enter? "))

lst = []
for i in range(n):
    element = input("Enter element: ")
    lst.append(element)

# Step 2: Take one more input to search and delete
item = input("Enter the element to delete: ")

if item in lst:
    lst.remove(item)
    print(item, "deleted from the list")
else:
    print(item, "not found in the list")

# Step 3: Iterate and print the updated list
print("Updated list:")
for i in lst:
    print(i)
```

```python
#You are given a list of integer elements. Make a new list that will store a square of elements of the previous list. (With and without list comprehension)
#i. Input_list = [2,5,6,12]
#ii. Output_list = [4,25,36,144]
input_list = [2, 5, 6, 12]

output_list = []

for num in input_list:
    output_list.append(num * num)

print(output_list)




```

```python
#without comprehension
input_list = [2, 5, 6, 12]

output_list = [num * num for num in input_list]

print(output_list)
```
